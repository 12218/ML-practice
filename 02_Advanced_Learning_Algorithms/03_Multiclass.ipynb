```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## 01 Prepare data

```python
# make 4-class dataset for classification
classes = 4
m = 100
centers = [[-5, 2], [-2, -2], [1, 2], [5, -2]]
std = 1.0
X_train, y_train = make_blobs(n_samples=m, centers=centers, cluster_std=std,random_state=30)
```

```python
print(f"X shape:{X_train.shape}", f"Y shape: {y_train.shape}")
print(f"Classes: {np.unique(y_train)}")
```

```python
c0_x = X_train[np.where(y_train == 0)]
c1_x = X_train[np.where(y_train == 1)]
c2_x = X_train[np.where(y_train == 2)]
c3_x = X_train[np.where(y_train == 3)]

c0_y = y_train[np.where(y_train == 0)]
c1_y = y_train[np.where(y_train == 1)]
c2_y = y_train[np.where(y_train == 2)]
c3_y = y_train[np.where(y_train == 3)]

plt.scatter(c0_x[:, 0], c0_x[:, 1], label = 'c0', color = 'red', marker = 'o', linewidths = 2)
plt.scatter(c1_x[:, 0], c1_x[:, 1], label = 'c1', color = 'blue', marker = 'o', linewidths = 2)
plt.scatter(c2_x[:, 0], c2_x[:, 1], label = 'c2', color = 'orange', marker = 'o', linewidths = 2)
plt.scatter(c3_x[:, 0], c3_x[:, 1], label = 'c3', color = 'purple', marker = 'o', linewidths = 2)

plt.xlabel('$X_0$')
plt.ylabel('$X_1$')
plt.title('Data set')

plt.legend()
plt.show()
```

## 02 Create Model

Notice the output layer uses a `linear` rather than a `softmax` activation. While it is possible to include the softmax in the output layer, it is <span style="color: red">more numerically stable</span> if linear outputs are passed to the loss function during training. If the model is used to predict probabilities, the softmax can be applied at that point.

```python
tf.random.set_seed(12218)

model = Sequential(
    [
        Dense(units = 2, activation = 'relu', name = 'layer1'),
        Dense(units = 4, activation = 'linear', name = 'layer2'),
    ]
)
```

## 03 Train

Setting `from_logits=True` as an argument to the loss function specifies that the output activation was linear rather than a softmax.

```python
model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    optimizer = tf.keras.optimizers.Adam(0.01)
)

model.fit(
    X_train,
    y_train,
    epochs = 200
)
```

## 04 Prediction

```python
prediction = model.predict(X_train)
```

```python
y_predict = np.argmax(prediction, axis = 1)
```
