```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_california_housing
```

## 01 Load data set

```python
ca_housing = fetch_california_housing()
print(ca_housing['DESCR'])
```

```python
ca_housing.keys()
```

```python
X_train = ca_housing.data[:300, :4] # get 300 instances from the data set
y_train = ca_housing.target[:300]
print(X_train.shape, y_train.shape)
```

## 02 Scale training data

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)
```

```python
print(f"Peak to Peak range by column in Raw        X:{np.ptp(X_train,axis=0)}")   
print(f"Peak to Peak range by column in Normalized X:{np.ptp(X_norm,axis=0)}")
```

### Show plot for normalised and unnormalised data set

```python
xaxis = np.arange(X_train.shape[0])
plt.figure()
plt.subplots_adjust(wspace = 1)
plt.subplot(1, 2, 1)
plt.title("Normalised data")
plt.xlabel("Normalised X_train")
plt.ylabel("y")
plt.scatter(xaxis, X_norm[:, 0], color = 'red', label = 'x1')
plt.scatter(xaxis, X_norm[:, 1], color = 'blue', label = 'x2')
plt.scatter(xaxis, X_norm[:, 2], color = 'yellow', label = 'x3')
plt.scatter(xaxis, X_norm[:, 3], color = 'green', label = 'x4')

plt.subplot(1, 2, 2)
plt.title("Unnormalised data")
plt.xlabel("Unnormalised X_train")
plt.ylabel("y")
plt.scatter(xaxis, X_train[:, 0], color = 'red', label = 'x1')
plt.scatter(xaxis, X_train[:, 1], color = 'blue', label = 'x2')
plt.scatter(xaxis, X_train[:, 2], color = 'yellow', label = 'x3')
plt.scatter(xaxis, X_train[:, 3], color = 'green', label = 'x4')

plt.legend()
plt.show()
```

## 03 Create and fit the regression model

### Create and fit model

```python
sgdr = SGDRegressor(max_iter = 1500)
sgdr.fit(X_norm, y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")
```

### View parameters

```python
b_norm = sgdr.intercept_
w_norm = sgdr.coef_
print(f"model parameters:\nw: {w_norm}, b:{b_norm}")
```

## 03 Make prediction

```python
y_prediction = sgdr.predict(ca_housing.data[300:350, :4])
print(f"Predict value: {y_prediction}")
print(f"Target value: {ca_housing.target[300:350]}")
```

## 04 Plot the prediction

```python
plt.figure()
plt.subplots_adjust(wspace = 1)

plt.subplot(1, 4, 1)
plt.xlabel(ca_housing.feature_names[0])
plt.ylabel("Price")
plt.scatter(ca_housing.data[300:350, 0], y_prediction, color = "blue", label = "Predict")
plt.scatter(ca_housing.data[300:350, 0], ca_housing.target[300:350], color = "orange", label = "Target")

plt.subplot(1, 4, 2)
plt.xlabel(ca_housing.feature_names[1])
plt.ylabel("Price")
plt.scatter(ca_housing.data[300:350, 1], y_prediction, color = "blue", label = "Predict")
plt.scatter(ca_housing.data[300:350, 1], ca_housing.target[300:350], color = "orange", label = "Target")

plt.subplot(1, 4, 3)
plt.xlabel(ca_housing.feature_names[2])
plt.ylabel("Price")
plt.scatter(ca_housing.data[300:350, 2], y_prediction, color = "blue", label = "Predict")
plt.scatter(ca_housing.data[300:350, 2], ca_housing.target[300:350], color = "orange", label = "Target")

plt.subplot(1, 4, 4)
plt.xlabel(ca_housing.feature_names[3])
plt.ylabel("Price")
plt.scatter(ca_housing.data[300:350, 3], y_prediction, color = "blue", label = "Predict")
plt.scatter(ca_housing.data[300:350, 3], ca_housing.target[300:350], color = "orange", label = "Target")

plt.legend()
plt.show()
```
