```python
x_train = np.array([1, 2])
y_train = np.array([300, 500])

print("Train set 01:")
print("Train X: {}".format(x_train))
print("Train Y: {}".format(y_train))
```

## 01 Computing cost

The equation for cost with one variable is:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 
  
 where 
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$
  
- $f_{w,b}(x^{(i)})$ is our prediction for example $i$ using parameters $w,b$.  
- $(f_{w,b}(x^{(i)}) -y^{(i)})^2$ is the squared difference between the target value and the prediction.   
- These differences are summed over all the $m$ examples and divided by `2m` to produce the cost, $J(w,b)$.

### (a) Computing $f(w, b)$

```python
def predict_y(w, b, x):
    return w * x + b
```

### (b) Computing $J(w, b)$

```python
def computing_cost(w, b, x, y):
    return np.sum((predict_y(w, b, x) - y) ** 2) / (2 * x.shape[0])
```

```python
print(computing_cost(220, 100, x_train, y_train))
```

### (c) Plot

```python
w = np.arange(0, 401)
b = 100

cost = np.zeros([401,])
for i in range(w.shape[0]):
    cost[i] = computing_cost(w[i], b, x_train, y_train)
    
mininum_cost = np.min(cost)
mininum_cost_index = np.argmin(cost)

print("Lowest cost: {}".format(mininum_cost))
print("w when cost is lowest: {}".format(mininum_cost_index))

plt.plot(w, cost, color = 'blue')
plt.plot(mininum_cost_index, mininum_cost, 'ro', label = 'Lowest cost')
plt.title("Cost for different w(b fixed at 100):")
plt.xlabel('w')
plt.ylabel('Cost')
plt.legend()
plt.show()
```

## 02 Gradient descent

In lecture, *gradient descent* was described as:

$$\begin{align*} \text{repeat}&\text{ until convergence:} \; \lbrace \newline
\;  w &= w -  \alpha \frac{\partial J(w,b)}{\partial w} \tag{3}  \; \newline 
 b &= b -  \alpha \frac{\partial J(w,b)}{\partial b}  \newline \rbrace
\end{align*}$$
where, parameters $w$, $b$ are updated **simultaneously**.  
The gradient is defined as:
$$
\begin{align}
\frac{\partial J(w,b)}{\partial w}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} ((f_{w,b}(x^{(i)}) - y^{(i)})x^{(i)}) \tag{4}\\
  \frac{\partial J(w,b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)}) \tag{5}\\
\end{align}
$$

Here *simultaniously* means that you calculate the partial derivatives for all the parameters before updating any of the parameters.

### (a) Compute $\frac{\partial (w, b)}{\partial w}$, $\frac{\partial (w, b)}{\partial b}$

```python
def compute_gradient(w, b, x, y):
    dj_dw = np.sum((predict_y(w, b, x) - y) * x) / x.shape[0]
    dj_db = np.sum(predict_y(w, b, x) - y) / x.shape[0]
    
    return dj_dw, dj_db
```

```python
print(x_train, y_train)
compute_gradient(0, 0, x_train, y_train)
```

### (b) Gradient descent

```python
def gradient_descent(x, y, w, b, alpha, num_iters):
    j_history = [] # history values of cost function
    p_history = [] # history values of w and b 
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(w, b, x, y)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        cost_n = computing_cost(w, b, x, y)
        
        if (i < 100000):
            j_history.append(cost_n)
            p_history.append([w, b])
            
        if (i % 1000 == 0):
            print(f"Iteration {i:4} - Cost: {cost_n:0.4e} - w: {dj_dw:0.4e} - b: {dj_db:0.4e}")
    return w, b, np.array(j_history), np.array(p_history)
```

### (c) Test gradient descent

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000
tmp_alpha = 1.0e-2
# run gradient descent
w_final, b_final, J_hist, p_hist = gradient_descent(x_train ,y_train, w_init, b_init, tmp_alpha, iterations)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

### (d) Plot for the first 100 iterations

```python
# xaxis = np.arange(1, len(J_hist) + 1)
xaxis = np.arange(1, 101)

plt.figure()
plt.subplots_adjust(hspace = 1)

plt.subplot(3, 1, 1)
plt.title("Cost after iterations")
plt.xlabel("Iteration step")
plt.ylabel("Cost")
plt.plot(xaxis, J_hist[:100], color = 'blue')

plt.subplot(3, 1, 2)
plt.title("w after iterations")
plt.xlabel("Iteration step")
plt.ylabel("w")
plt.plot(xaxis, p_hist[:100, 0], color = 'blue')

plt.subplot(3, 1, 3)
plt.title("b after iterations")
plt.xlabel("Iteration step")
plt.ylabel("b")
plt.plot(xaxis, p_hist[:100, 1], color = 'blue')

plt.show()
```

## 02 Problem Statement

You will use the motivating example of housing price prediction. The training dataset contains three examples with four features (size, bedrooms, floors and, age) shown in the table below.  Note that, unlike the earlier labs, size is in sqft rather than 1000 sqft. This causes an issue, which you will solve in the next lab!

| Size (sqft) | Number of Bedrooms  | Number of floors | Age of  Home | Price (1000s dollars)  |   
| ----------------| ------------------- |----------------- |--------------|-------------- |  
| 2104            | 5                   | 1                | 45           | 460           |  
| 1416            | 3                   | 2                | 40           | 232           |  
| 852             | 2                   | 1                | 35           | 178           |  

You will build a linear regression model using these values so you can then predict the price for other houses. For example, a house with 1200 sqft, 3 bedrooms, 1 floor, 40 years old.  

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

$\mathbf{w}$ is a vector with $n$ elements.

$$\mathbf{w} = \begin{pmatrix}
w_0 \\ 
w_1 \\
\cdots\\
w_{n-1}
\end{pmatrix}
$$
$b$ is a scalar parameter.

```python
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
print(f"w_init shape: {w_init.shape}, b_init type: {type(b_init)}")
```

**For multiple-variable prediction, we should use dot product of vectors.**

```python
np.dot(X_train, w_init)
```

### (a) Computing $f(w, b)$

```python
def predict_y_multiple(w, b, x): # w and x are both vectors
    return np.dot(x, w) + b
```

```python
predict_y_multiple(w_init, b_init, X_train[0, :])
```

### (b) Computing $J(w, b)$

```python
def computing_cost_multiple(w, b, x, y):
    return np.sum((predict_y_multiple(w, b, x) - y) ** 2) / (2 * x.shape[0])
```

```python
computing_cost_multiple(w_init, b_init, X_train, y_train)
```

### (c) Compute $\frac{\partial (w, b)}{\partial w}$, $\frac{\partial (w, b)}{\partial b}$

```python
def compute_gradient_multiple(w, b, x, y):
    dj_dw = 0
    for i in range(X_train.shape[0]):
        dj_dw_i = (predict_y_multiple(w, b, x[i]) - y[i]) * x[i] / x.shape[0]
        dj_dw += dj_dw_i
    dj_db = np.sum(predict_y_multiple(w, b, x) - y) / x.shape[0]
    
    return dj_dw, dj_db
compute_gradient_multiple(w_init, b_init, X_train, y_train)
```

```python
#Compute and display gradient 
tmp_dj_dw, tmp_dj_db = compute_gradient_multiple(w_init, b_init, X_train, y_train)
print(f'dj_db at initial w,b: {tmp_dj_db}')
print(f'dj_dw at initial w,b: \n {tmp_dj_dw}')
```

### (d) Gradient descent

```python
def gradient_descent_multiple(w, b, x, y, alpha, num_iters):
    j_history = [] # history values of cost function
    w_history = [] # history values of w
    b_history = [] # history values of b
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient_multiple(w, b, x, y)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        cost_n = computing_cost_multiple(w, b, x, y)
        
        if (i < 100000):
            j_history.append(cost_n)
            w_history.append(w)
            b_history.append(b)
            
        if (i % 100 == 0):
            print(f"Iteration {i:4}: Cost: {cost_n:0.4e} - w: {w} - b: {b}")
    return w, b, j_history, w_history, b_history
```

```python
# initialize parameters
initial_w = np.zeros_like(w_init)
initial_b = 0.
# some gradient descent settings
iterations = 1000
alpha = 5.0e-7
# run gradient descent 
w_final, b_final, J_hist, w_hist, b_hist = gradient_descent_multiple(initial_w, initial_b, X_train, y_train, alpha, iterations)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
m,_ = X_train.shape
for i in range(m):
    print(f"prediction: {np.dot(X_train[i], w_final) + b_final:0.2f}, target value: {y_train[i]}")
```

### (e) Plot of gradient descent

```python
# xaxis = np.arange(1, len(J_hist) + 1)
xaxis = np.arange(1, 101)

plt.figure()

plt.title("Cost after iterations")
plt.xlabel("Iteration step")
plt.ylabel("Cost")
plt.plot(xaxis, J_hist[:100], color = 'blue')

plt.show()
```
